```python
def fibonacci(max):        # генератор (а не функция, т.к. оператор return заменён на yield)
    a, b = 0, 1
    while a < max:
        yield a            # return a, + запоминаем место рестарта для следующего вызова
        a, b = b, a + b    # параллельное присваивание, которое выполняется одновременно и параллельно

for n in fibonacci(100):   # используем генератор fibonacci() как итератор
    print(n)               # печатаем все числа Фибоначчи меньшие 100 через пробел
```

# Пример математической формулы

Вот пример квадратного уравнения:

$$ ax^2 + bx + c = 0 $$

Где:
- \( a \), \( b \), и \( c \) - это коэффициенты,
- \( x \) - переменная.

Решение квадратного уравнения находится по формуле:

$$ x = \frac{{-b \pm \sqrt{{b^2 - 4ac}}}}{{2a}} $$

Также можно написать и более сложные уравнения, например, формулу суммы ряда:

$$ S = \sum_{n=1}^{\infty} \frac{1}{n^2} = \frac{\pi^2}{6} $$


```python
# Импорт необходимых библиотек
import matplotlib.pyplot as plt
import numpy as np

# Данные для графика
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Создание графика
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)')

# Добавление заголовка и меток осей
plt.title('График синусоиды')
plt.xlabel('x')
plt.ylabel('sin(x)')

# Добавление легенды
plt.legend()

# Отображение графика
plt.grid(True)
plt.show()
```
